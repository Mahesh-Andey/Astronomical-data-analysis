Viewing Image with Different Color Map

```python
import matplotlib.pyplot as plt
```

```python
img_gray = io.imread("/content/Capture.PNG", as_gray=True)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="jet")
```

```python
plt.imshow(img_gray, cmap="hot")
```

```python
#Multiple plots using pyplot
fig = plt.figure(figsize=(20, 25))

ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_gray, cmap='hot')
ax1.title.set_text('1st')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(img_gray, cmap='jet')
ax2.title.set_text('2nd')

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(img_gray, cmap='magma')
ax3.title.set_text('3rd')

ax4 = fig.add_subplot(2,2,4)
ax4.imshow(img_gray, cmap='nipy_spectral')
ax4.title.set_text('4th')
plt.show()
```

Resize and Rescale 

```python
from skimage import io, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
img = io.imread("/content/Capture.PNG", as_gray=True)

plt.imshow(img)
```

```python
img
```

```python
#gaussian smoothing can performed to avoid anti aliasing artifacts.
img_rescaled = rescale(img, 1.0 / 4.0,)  #Check rescales image size in variable explorer

plt.imshow(img_rescaled)
```

```python
img_rescaled
```

```python
img_rescaled.shape
```

```python
#Resize, resize image to given dimensions (shape)
img_resized = resize(img, (200, 200) )

plt.imshow(img_resized)
```

```python
img_resized
```

```python
img_resized.shape
```

```python
#Downscale, downsample using local mean of elements of each block defined by user
img_downscaled = downscale_local_mean(img, (4, 3))                  # 348/4, 679/3

plt.imshow(img_downscaled)
```

```python
img_downscaled
```

```python
img_downscaled.shape
```
